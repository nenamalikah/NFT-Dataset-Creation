```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://api.coinmarketcap.com/data-api/v3/nft/collections?start=0&limit=100&sort=volume&desc=true&period=4"

payload={}
headers = {
  'Accept': 'application/json, text/plain, */*',
  'Origin': 'https://coinmarketcap.com',
  'Referer': 'https://coinmarketcap.com/',
  'Host': 'api.coinmarketcap.com',
  'Accept-Language': 'en-US,en;q=0.9',
  'User-Agent': '',
  'Accept-Encoding': 'gzip, deflate, br',
  'Connection': 'keep-alive',
  'FVIDEO-ID': '',
  'x-request-id': ''
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

```python
values = response.json()
```

```python
def webpage_to_df(url, headers):
    
    payload={}
    
    requested = requests.request("GET", url, headers=headers, data=payload)
    
    web_dict = requested.json()
    
    web_text = web_dict['data']['collections']
    
    returned_df = pd.DataFrame.from_dict(web_text, orient='columns')
    
    return returned_df
```

```python
for_join = [df, df2, df3, df4, df5, df6]
```

```python
table = pd.concat(for_join)
table.head()
```

```python
table_f = table.drop(columns=['slug', 'dataSource','blockchain', 'oneDay', 'sevenDay', 'thirtyDay', 
                    'volume7d', 'sales7d'])
table_f.head()
```

```python
table_f.drop(columns=['rank', 'popular'], inplace=True)
```

```python
Average_Price = []
Volume = []

for x in table_f['allTime']:
    Average_Price.append(x['averagePrice'])
    Volume.append(x['volume'])
    
table_f['Average_Price'] = Average_Price
table_f['Volume'] = Volume

table_f.head()
```

```python
Category = []

for x in table_f['categories']:
    try:
        Category.append(','.join(x))
    except:
        Category.append(None)

table_f['Category'] = Category
table_f.head()
```

```python
table_f.drop(columns=['categories', 'allTime', 'floorPriceToken'], inplace=True)
```

```python
table_f.rename(columns={'name':'Name', 'floorPrice':'Floor_Price', 'logo':'Logo',
                       'website':'Website', 'marketCap':'Market_Cap', 'owners':'Owners',
                       'assets':'Assets', 'ownerAssetsPercentage':'Owner_Asset_Ratio',
                       'netWorth':'Net_Worth', 'salesAT':'Sales'}, inplace=True)
```

```python
column_names = ['Name', 'Volume', 'Market_Cap', 'Sales', 'Floor_Price', 'Average_Price',
               'Owners', 'Assets', 'Owner_Asset_Ratio', 'Net_Worth', 'Category', 'Website', 'Logo']

df_nft = table_f.reindex(columns=column_names)
df_nft.head()
```

```python
Volume_USD = []
Market_Cap_USD = []
Floor_Price_USD = []
Average_Price_USD = []
```

```python
for x in df_nft['Volume']:
    Volume_USD.append(x*146.82)
    
df_nft['Volume_USD'] = Volume_USD
```

```python
for x in df_nft['Market_Cap']:
    Market_Cap_USD.append(x*146.82)
    
df_nft['Market_Cap_USD'] = Market_Cap_USD
```

```python
for x in df_nft['Floor_Price']:
    Floor_Price_USD.append(x*146.82)
    
df_nft['Floor_Price_USD'] = Floor_Price_USD
```

```python
for x in df_nft['Average_Price']:
    Average_Price_USD.append(x*146.82)
    
df_nft['Average_Price_USD'] = Average_Price_USD
```

```python
column_names = ['Name', 'Volume', 'Volume_USD', 'Market_Cap', 'Market_Cap_USD', 
                'Sales', 'Floor_Price', 'Floor_Price_USD', 'Average_Price', 'Average_Price_USD', 
                'Owners', 'Assets', 'Owner_Asset_Ratio', 'Net_Worth', 'Category', 'Website', 'Logo']

df_nft = df_nft.reindex(columns=column_names)
df_nft.head()
```
